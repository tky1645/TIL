```python
from transformers import pipeline
```

```python
text_classification_pipeline = pipeline(model="llm-book/bert-base-japanese-v3-marc_ja")
positive_text = "世界には言葉がわからなくても感動する音楽がある。"
# positive_textの極性を予測
print(text_classification_pipeline(positive_text)[0])
```

```python
# negative_textの極性を予測
negative_text = "世界には言葉がでないほどひどい音楽など存在しない。"
print(text_classification_pipeline(negative_text)[0])
```

意味的類似度計算

```python
text_sim_pipeline = pipeline(
     model="llm-book/bert-base-japanese-v3-jsts",
     function_to_apply="none",
 )
text = "自ら設計、計画したものを世の中に出すことによって社会に貢献したいからです。特に私は貴社の建物トータルソリューションに興味があり、その中でより省エネルギーを目指すとともに利用者にも快適に過ごしてもらえるような環境を提供できるような提案をしていきたいと考えているからです。そうすることによってビルの利用者の方たちにも快適で安心できる空間を提供しつつ、実際の電力使用量を抑えられことによりCO2排出量の削減にもつながり地球環境にもつなげられると考えました。そのような総合的な提案や設計、計画に携われる職種を考えたときに総合職であると考えたので志望しました。"
sim_text = "自動車が好きだからです。私は自動車に関わるものづくりを行っていきたいと考えています。ですから自分で開発設計した製品により世の中の自動車での電気の使用効率を向上させ環境性・経済性を良くすることで社会や人々に貢献したいです。今後電気自動車やハイブリッド自動車の普及が進んでいき、そのため貴社の自動車の製品の需要、種類は大きくなっていくと思います。また将来的に電気自動車を普及させるためのインフラ設備も重要になってきます。ですから自動車の充電スタンドにも携わり電気自動車の電気残量を気にしなくてもよいような社会にして多くの人に自動車で快適に走る喜びを感じてもらいたいと思いました。ですから私も貴社の中で専門としてきた電気の知識を活用して製品の開発設計を行いより自動車が効率よく電気エネルギーを使用できるような製品を製作したいので志望しました。"
# textとsim_textの類似度を計算
result = text_sim_pipeline({"text": text, "text_pair": sim_text})
print(result["score"])
```
